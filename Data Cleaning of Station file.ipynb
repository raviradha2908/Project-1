```python
 #this intiates the context based, automatically closes after its block finishes
#open action with path to JSON file --> utf-8 ensures reading of non ASCII characters
#as f ensures alias -> which can be used for loading

with open("C:/Users/ravir/Project 1/stations.json",encoding="utf-8") as f:
    data = json.load(f)
    
#data.keys() --> used to display major sections in JSON file
#output has types and features --> type describes overall structure and features has list of items
#in this no metadata is present
#normalise function helps in flattening nested dictionaries in JSON like data structures into tabular dataframe

JSON= pd.json_normalize(data["features"], sep = "_",max_level = 2) #here underscores as separator and flattens only upto 2 layers
JSON.to_csv("Station_data.csv",index = False)
```

```python
df = pd.read_csv("C:/Users/ravir/Project 1/Station_data.csv")
df.head()
```

```python
df.dropna(subset=["geometry_coordinates","properties_name"],inplace = True)
df.drop(columns =["type","geometry_type","properties_zone","geometry"], axis = 1,inplace =True)
```

```python
df["geometry_coordinates"]= df["geometry_coordinates"].str.strip("[]")
df[["Latitudes","Longitudes"]] = df["geometry_coordinates"].str.split(",",expand = True)
df["Latitudes"] = df["Latitudes"].str.strip("[")
df["Longitudes"] =df["Longitudes"].str.strip("]")
df=df.drop("geometry_coordinates",axis =1 )
df["Latitudes"]= df["Latitudes"].astype(float)
df["Longitudes"]= df["Longitudes"].astype(float)
df
```

```python
df.columns = ["Station State","Station Code","Station Name","Station Address","Latitudes","Longitudes"]
df
```

```python
import pdfplumber
pdf_path ="C:/Users/ravir/Project 1/Railway_station_zone-category_wise_list.pdf"
station_data_with_state= []
with pdfplumber.open(pdf_path) as station_data_with_state_pdf:
    for page_num, page in enumerate(station_data_with_state_pdf.pages):
        table = page.extract_table()
        if table:
            for row in table[:]:
                station_data_with_state.append(row)
```

```python
station_data_with_state = pd.DataFrame(station_data_with_state)
station_data_with_state.columns = station_data_with_state.iloc[0]
station_data_with_state = station_data_with_state.drop(station_data_with_state.index[0])
station_data_with_state.reset_index(drop = True,inplace = True)
station_data_with_state
```

```python
station_data_with_state["Station Code"] = station_data_with_state["Station Code"].str.upper()
df["Station Code"] = df["Station Code"].str.upper()
station_merged = pd.merge(station_data_with_state, df[["Station Code","Latitudes","Longitudes"]], on = "Station Code", how = "left")
station_merged.drop("S No", axis =1, inplace = True)
```

```python
station_merged.dropna(subset ="Latitudes", inplace = True)
station_merged.to_csv("Station_refined.csv", index = False)
```
